```python
#taks 5 *HARD*
def change_columns(name):
    new_name = name.lower().replace(' ', '_')
    return new_name

bookings = bookings.rename(columns = change_columns)
```

```python
#task 6 *HARD*
bookings.query('is_canceled == 0').groupby('country').agg({"is_canceled":"count"}).sort_values("is_canceled",ascending=False).head(5)
```

```python
#task 10
bookings.query('is_canceled == 1').groupby(['arrival_date_year','arrival_date_month']).agg({'is_canceled':'count'})
```

```python
#task 11
bookings.agg({"adults" : "mean", "children" : "mean", "babies" : "mean"})
```

```python
#task 12
children_plus_babies = bookings['children'] + bookings['babies']
bookings['total_kids'] = children_plus_babies
bookings.groupby('hotel').agg({'total_kids' : 'mean'}).round(2)
```

```python
#task 13 *HARD xD*
# если есть хотяб один ребенок, то True else False
has_kids = bookings['total_kids'] >= 1
# создание столбца
bookings['has_kids'] = has_kids
# с детьит и отменили
has_c = bookings.query('has_kids == True and is_canceled == 1').agg({'is_canceled': 'count'})
# все с детьми
hass  = bookings.query('has_kids == True').agg({'is_canceled': 'count'})
# без детей и отменили
hasnt_c = bookings.query('has_kids == False and is_canceled == 1').agg({'is_canceled': 'count'})
# все без детей
hasntt  = bookings.query('has_kids == False').agg({'is_canceled': 'count'})
#total_c = bookings.query('is_canceled == 1').agg({'total_c': 'sum'})
# Churn rate с детьми. С детьит и отменили / все с детьми   
toto = (has_c/hass)
# Churn rate Без детей. без детей и отменили / все без детей
nc_c = (hasnt_c/hasntt)
```
