```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
mu = 100
sigma = 10
```

```python
population = np.round(np.random.normal(mu,sigma,10000),2)
```

```python
population.std()
```

```python
sns.distplot(population, kde=False)
```

```python
sample_size = 30
```

```python
sample = np.random.normal(mu,sigma,10)
```

```python
sample.mean()
```

```python
sample = np.random.choice(population,sample_size,False)
```

```python
sample.mean()
```

```python
sample_means = []
```

```python
n = 1000
```

```python
for i in range(n):
    sample = np.random.choice(population,sample_size,False)
    sample_means.append(sample.mean())
```

```python
sample_means
```

```python
sns.distplot(sample_means,kde=False)
```

```python
se = population.std() / np.sqrt(sample_size)
se
```

```python
np.std(sample_means)
```
