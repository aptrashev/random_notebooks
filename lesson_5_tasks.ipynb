```python
#task 5
gt = retail.query('Country == "Germany"').groupby('CustomerID',as_index=False).agg({'InvoiceNo':'nunique'})
```

```python
gt.InvoiceNo.quantile(q=0.80)
```

```python
#task 6
top_retail_germany  = retail.query('CustomerID in @germany_top') 
```
