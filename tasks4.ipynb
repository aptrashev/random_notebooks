```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
taxi = pd.read_csv('./3_taxi_peru.csv', sep=';', parse_dates=['start_at', 'end_at', 'arrived_at'])
```

```python
taxi
```

```python
taxi['wait_time'] = (taxi.arrived_at - taxi.start_at).astype('timedelta64[m]')
```

```python
taxi.head()
```

```python
taxi.query('wait_time > 0.0 and start_type == "reserved"').groupby('driver_id').agg({'wait_time':'count'}).sort_values('wait_time',ascending=False)
```

```python
taxi.query('driver_id == "406921adcca37705ef527b4246c0cfea"')
```

```python
taxi['month'] = (taxi.start_at).dt.month
```

```python
taxi.head()
```

```python
taxi['weekday'] = (taxi.start_at).dt.strftime('%A')
```

```python
taxi.head()
```

```python
ord = taxi.groupby('month').agg({'journey_id' : 'count'})
```

```python
ord
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=taxi,x='weekday',order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
```

```python
mau = taxi.groupby('month').agg({'user_id': 'nunique' })
```
